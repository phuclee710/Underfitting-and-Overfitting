## 1. Overfitting and Underfitting

 Any model when trained inefficiently can lead to two bad effects called Overfitting and Underfitting. Overfitting is when the model we built perfoms overwhelmingly on the training set but fails to capture the correct predictions on the test set. Underfitting is when  the model fails to fit the training data and performs badly on the test data as well. By dimensionality reduction, feature selection and proper training one can get better of this issue.

The first step is to split the data into training and test sets:

```python
## Splitting the data 
def split_data(data,prob)
    results=[],[]
    for row in data:
        results[0 if random.random()<prob else 1].append(row)
    return results
```

When we have separate matrices x and y for input variables and output respectively, we need to make sure that they are together in the training and the test data sets.

```python
def test_train_split(x,y,pct)
    data=zip(x,y)
    train,test=split_data(data,1-pct)
    x_train,y_train= (*train) #un-zip trick
    x_test, y_test= (*test)
    return(x_train,y_train,x_test,y_test)
```

But, if there are common patterns in train and test data that wouldn't generalize for an outside large data set; this method might fail.
In such case, data should be split into three parts-
- Test Set
- Validation Set 
- Training Set

## 2. Bias- Variance Trade-off

It's an another way of thinking about overfitting and underfitting problem.
- For any underfitting model, when two random data sets are choosen from the training set, they approximately fit the model in the same manner and the variance is low. But each of the training set has a lot of errors because of underfitting. Hence an under fitting model has high bias and low variance.
- For an overfitting model, it will the case completely opposite to as said above. It has low bias and high variance.

If the model has high bias, more features should be added to fit the data to the model and if the model gas high variance, features should be removed or more data should be gathered.
